## Read-in parlInfo csv file

```python
#Used to obtain most relevant information

init_df = pd.read_csv("Senate_Hansard_Parliamentarians.csv",delimiter=",")

init_df
```

```python
#Get list of indigenous senators

init_df_ind = pd.read_csv("Indigenous_Parliamentarians.csv",delimiter=",")

init_df_ind
```

```python
#Convert data from parlinfo to applicable columns in new authorityfile dataframe

#Get first, middle and last names, also checks if senators have them in first place, otherwise empty
fullName = init_df["Name"]
lastName = [namestring.split(", ")[0] if "," in namestring else "" for namestring in init_df["Name"]]
firstmiddleName = [namestring.split(", ")[1] if "," in namestring else namestring for namestring in init_df["Name"]]
firstName, middleName = [namestring.split(" ")[0] if " " in namestring else namestring for namestring in firstmiddleName], [namestring.split(" ")[1] if " " in namestring else "" for namestring in firstmiddleName]

#print(lastName)
#print(firstName)
#print(middleName)

sex = [genderstring.lower() for genderstring in init_df["Gender"]]
#print(sex)

indigenous = [1 if namestring in list(init_df_ind["Name"]) else 0 for namestring in init_df["Name"]]
#print(indigenous)

dateOfBirth = [datetime.strptime(DOBstring, "%Y-%m-%d") if "-" in DOBstring else "" for DOBstring in init_df["Date of Birth"]]
#print(dateOfBirth)

isEstimateDOB = [0 for datestring in dateOfBirth]

#NOTE: \xa0 represents a space in HTML syntax
birthProvince = ["" if birthprovince == "\xa0" else birthprovince for birthprovince in init_df["Province/Region of Birth"]]
#print(birthProvince)

birthCountry = ["" if birthcountry == "\xa0" else birthcountry for birthcountry in init_df["Country of Birth"]]
#print(birthCountry)

#Convert init_df's Type of Parliamentarian category to first day in correct format
def firstdayfunct(x):
    if "Senator" in x:
        x = x.split("Senator (")[1]
    else:
        ""
    x = x.split(" -")[0]
    x = x.replace("/","-")
    return x

firstDay = [datetime.strptime(firstdayfunct(x), "%Y-%m-%d") for x in init_df["Type of Parliamentarian"]]
#print(firstDay)

provOfRiding = [province for province in init_df["Province/Territory"]]
#print(provofRiding)
```
